# Explanation



| Result  | Meaning                        | Explanation                                                             |
| ------- | ------------------------------ | ----------------------------------------------------------------------- |
| `False` | No protected material detected | The text does not contain copyrighted, duplicated, or protected content |
| `True`  | Protected material detected    | The system identified potentially copyrighted or protected text content |


# Sample 1

question 1 : What is the step by step guide to invest in share market in india?

```python
#
# Copyright (c) Microsoft. All rights reserved.
# To learn more, please visit the documentation - Quickstart: Azure Content Safety: https://aka.ms/acsstudiodoc
#

import requests

# Replace with your own subscription_key and endpoint
subscription_key = "#"
endpoint = "https://projectcontent.cognitiveservices.azure.com/"


# The text to be analyzed
text_to_analyze = "What is the step by step guide to invest in share market in india?"

# Set up the API request
url = f"{endpoint}/contentsafety/text:detectProtectedMaterial?api-version=2024-09-01"
headers = {
  "Content-Type": "application/json",
  "Ocp-Apim-Subscription-Key": subscription_key
}
data = {
  "text": text_to_analyze
}

# Send the API request
response = requests.post(url, headers=headers, json=data)

# Handle the API response
if response.status_code == 200:
    result = response.json()
    print("Analysis result:", result)
else:
    print("Error:", response.status_code, response.text)
```

```python
df = df.head(30)
```

Using both columns

```python
# API URL
url = f"{endpoint}/contentsafety/text:detectProtectedMaterial?api-version=2024-09-01"

# Headers
headers = {
    "Content-Type": "application/json",
    "Ocp-Apim-Subscription-Key": subscription_key
}

# Loop through rows
for index, row in df.iterrows():

    question1 = str(row["question1"])
    question2 = str(row["question2"])

    print(f"\nRow {index}")

    # Analyze Question 1
    data_q1 = {
        "text": question1
    }

    response_q1 = requests.post(url, headers=headers, json=data_q1)

    if response_q1.status_code == 200:
        result_q1 = response_q1.json()
        print("Question 1:")
        print(question1)
        print("Analysis result:", result_q1)
    else:
        print("Error Q1:", response_q1.status_code, response_q1.text)

    # Analyze Question 2
    data_q2 = {
        "text": question2
    }

    response_q2 = requests.post(url, headers=headers, json=data_q2)

    if response_q2.status_code == 200:
        result_q2 = response_q2.json()
        print("\nQuestion 2:")
        print(question2)
        print("Analysis result:", result_q2)
    else:
        print("Error Q2:", response_q2.status_code, response_q2.text)
```

- The Protected Material Detection analysis performed using Microsoft Azure AI showed that all analyzed text inputs returned a detection result of False, indicating that no protected, copyrighted, or duplicated material was identified in the dataset.

- Based on the analysis results, all question pairs were classified as safe and acceptable because the system did not detect any protected content violations. The analyzed text did not contain copyrighted articles, protected documentation, song lyrics, movie dialogue, or other restricted textual materials.
