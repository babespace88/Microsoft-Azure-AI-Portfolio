# DEPLOY AZURE AI INVOICE FRAUD DETECTION AGENT

```python
# =========================================
# INSTALL
# pip install azure-ai-projects>=2.1.0
# pip install azure-identity
# =========================================

from azure.identity import DefaultAzureCredential
from azure.ai.projects import AIProjectClient
import base64

# =========================================
# AZURE AI FOUNDRY CONFIG
# =========================================

endpoint = "#"

project_client = AIProjectClient(
    endpoint=endpoint,
    credential=DefaultAzureCredential(),
)

# =========================================
# AGENT INFO
# =========================================

my_agent = "ai-detection-invoice"
my_version = "8"

# =========================================
# GET OPENAI CLIENT
# =========================================

openai_client = project_client.get_openai_client()

# =========================================
# LOAD IMAGE
# =========================================

image_path = "/content/batch1-0001.jpg"

with open(image_path, "rb") as image_file:
    base64_image = base64.b64encode(
        image_file.read()
    ).decode("utf-8")

# =========================================
# AI DETECTION PROMPT
# =========================================

prompt = """
You are an AI invoice fraud detection assistant.

Analyze the uploaded invoice image and compare it with known invoice patterns.

Your task:
- Determine whether the invoice is REAL or FAKE
- Explain briefly why
- Keep the answer short and professional
- Do not return JSON
- Do not explain technical details

Example response:
"This invoice is likely FAKE because the invoice amount appears suspicious and the structure differs from known invoices."
"""

# =========================================
# SEND REQUEST TO AGENT
# =========================================

response = openai_client.responses.create(

    model="gpt-4.1",

    input=[
        {
            "role": "user",
            "content": [

                {
                    "type": "input_text",
                    "text": prompt
                },

                {
                    "type": "input_image",
                    "image_url": f"data:image/jpeg;base64,{base64_image}"
                }

            ]
        }
    ],

    extra_body={
        "agent_reference": {
            "name": my_agent,
            "version": my_version,
            "type": "agent_reference"
        }
    }

)

# =========================================
# OUTPUT
# =========================================

print("\n========== AI INVOICE DETECTION ==========\n")

print(response.output_text)
```
